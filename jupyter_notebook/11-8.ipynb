# 넘파이, 판다스, 매트플랏립 실습

```python
# 그래프를 바로 보기 위한 명령어
%matplotlib inline
import numpy as np # 넘파이 탑재
import pandas as pd # 판다스 탑재
import matplotlib # 매트플랏립 탑재
```

```python
# 50개의 난수로 이루어진 데이터 생성
data = np.random.rand(50)
data
```

```python
# 판다스 시리즈 타입 데이터로 변환
seri = pd.Series(data)
seri
```

```python
# 선형 그래프 그리기
seri.plot()
```
